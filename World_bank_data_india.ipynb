```
import seaborn as sns
import plotly.io as pio
pio.templates.default = "plotly_dark"
sns.set(style="darkgrid")
```

```
import plotly.graph_objects as go
```

```

trace1 = go.Scatter(
                    x = df.GDP_IND,
                    y = df.EMP_AGR,
                    mode = "lines",
                    name = "EMP",
                    marker = dict(color = 'rgba(16, 112, 2, 0.8)'),
                    text= df.EMP_TOTL
                    )
# Creating trace2
trace2 = go.Scatter(
                    x = df.GDP_IND,
                    y = df.GDP_AGR,
                    mode = "lines+markers",
                    name = "GDP",
                    marker = dict(color = 'rgba(80, 26, 80, 0.8)'),
                    text= df.EMP_TOTL)
data = [trace1, trace2]
layout = dict(title = 'EMP vs GDP',
              xaxis= dict(title= 'EMP',ticklen= 5,zeroline= False)
             )
fig = dict(data = data, layout = layout)
plt.show(fig)
```

```
import plotly.express as px
df=pd.read_csv('World_Bank_Data_India.csv')
fig = px.bar_polar(df, r='POP_TOTL', theta="EMP_TOTL",
                   color="Years", template="plotly_dark",
                   color_discrete_sequence= px.colors.sequential.Plasma_r)
fig.show()
```
